```python
import matplotlib.pyplot as plt
import numpy as np
import folium

# Example coordinates for locations (replace with real coordinates if needed)
location_coordinates = {
    0: (0, 0),   # Depot
    1: (10, 20), # Location 1
    2: (20, 10), # Location 2
    3: (30, 40), # Location 3
    4: (40, 30)  # Location 4
}

# Example routes from the VRP solver (replace with actual data)
example_routes = {
    1: [(0, 0, 10, 5, 100), (1, 10, 20, 3, 200)],  # Vehicle 1
    2: [(0, 0, 10, 5, 100), (2, 20, 30, 4, 150), (3, 30, 40, 2, 250)]  # Vehicle 2
}

def plot_vrp_routes(routes, location_coords, use_folium=False):
    """ Plots vehicle routes using Matplotlib or Folium. """
    
    colors = ['blue', 'orange', 'green', 'purple', 'brown']  # Colors for vehicles
    depot_x, depot_y = location_coords[0]  # Depot location

    if use_folium:
        # Initialize a folium map centered at the depot
        m = folium.Map(location=[depot_y, depot_x], zoom_start=12)
        
        # Add depot marker
        folium.Marker([depot_y, depot_x], popup="Depot", icon=folium.Icon(color="red")).add_to(m)

        # Plot each vehicle's route
        for vehicle_id, route in routes.items():
            route_coords = [(location_coords[stop[0]][1], location_coords[stop[0]][0]) for stop in route]
            folium.PolyLine(route_coords, color=colors[vehicle_id % len(colors)], weight=3, opacity=0.8).add_to(m)
            
            # Add markers with labels
            for stop in route:
                loc_id, arrival, departure, load, distance = stop
                lat, lon = location_coords[loc_id]
                folium.Marker(
                    [lon, lat], 
                    popup=f"Loc: {loc_id}, Arrival: {arrival}, Load: {load}"
                ).add_to(m)
        
        return m  # Return folium map object
    
    else:
        # Matplotlib visualization
        plt.figure(figsize=(10, 6))

        for vehicle_id, route in routes.items():
            x_coords = []
            y_coords = []

            for stop in route:
                loc_id, arrival, departure, load, distance = stop
                x, y = location_coords[loc_id]
                x_coords.append(x)
                y_coords.append(y)
                
                # Add labels for each stop
                plt.text(x, y, f"{loc_id}\n({arrival})", fontsize=9, ha='right', va='bottom', color='black')

            # Close loop back to depot
            x_coords = [depot_x] + x_coords + [depot_x]
            y_coords = [depot_y] + y_coords + [depot_y]

            # Plot routes
            plt.plot(x_coords, y_coords, marker='o', color=colors[vehicle_id % len(colors)], label=f'Vehicle {vehicle_id}')

            # Add arrows for direction
            U = np.diff(x_coords)
            V = np.diff(y_coords)
            X_mid = x_coords[:-1]
            Y_mid = y_coords[:-1]
            plt.quiver(X_mid, Y_mid, U, V, angles='xy', scale_units='xy', scale=1, color='black', alpha=0.7, width=0.005)

        # Plot depot
        plt.scatter(depot_x, depot_y, color='red', label='Depot', s=100, zorder=5)

        # Add title and legend
        plt.title('Vehicle Routes with Labels & Arrows')
        plt.xlabel('X Coordinate')
        plt.ylabel('Y Coordinate')
        plt.legend()
        plt.grid(True)
        plt.show()

# **Call the function with use_folium=False for Matplotlib, True for Folium**
plot_vrp_routes(example_routes, location_coordinates, use_folium=True)
```
