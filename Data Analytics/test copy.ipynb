```python
def determine_century(mmddyy, pivot_year=24):
    """
    Determines the full year and century for a given MMDDYY format date.
    
    Args:
    mmddyy (str): Date in MMDDYY format (e.g., "031924").
    pivot_year (int): The pivot year to determine the century (default is 24).
    
    Returns:
    int: The full year (e.g., 1924 or 2024).
    """
    # Extract the year (last two digits)
    yy = int(mmddyy[-2:])
    
    # Determine the century
    if yy <= pivot_year:
        century = 2000  # Belongs to 2000s
    else:
        century = 1900  # Belongs to 1900s
    
    # Combine century and year
    full_year = century + yy
    return full_year

# Examples
print(determine_century("012424"))  # Output: 2024
print(determine_century("031924"))  # Output: 1924
```
