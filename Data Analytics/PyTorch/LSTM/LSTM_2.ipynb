```python
import numpy as np
import torch
import torch.nn as nn

from sklearn.preprocessing import MinMaxScaler

import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data = np.sin(np.linspace(0, 100, 500)) + np.random.normal(0, 0.1, 500)
data = data.reshape(-1, 1)
```

```python
data.shape
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)
```

```python
def create_sequences(data, sequence_length):
    sequences = []
    for i in range(len(data)-sequence_length):
        seq = data[i:i+sequence_length]
        label = data[i+sequence_length]
        sequences.append((seq, label))
    return sequences
```

```python
sequence_length = 5
sequences = create_sequences(data=data_scaled, sequence_length=sequence_length)
```

```python
sequences
```

```python
X = np.array([seq[0] for seq in sequences])
y = np.array([seq[1] for seq in sequences])
```

```python
Z = np.array()
```

```python
Z
```

```python
X
```

```python
X.shape
```

```python
X_tensor = torch.FloatTensor(X)
y_tensor = torch.FloatTensor(y)
```

```python
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super().__init__()
        self.hidden_size = hidden_size
        self.lstm = nn.LSTM(input_size=input_size,
                            hidden_size=hidden_size,
                            batch_first=True
                            )
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        _, (hidden, _) = self.lstm(x)
        out = self.fc(hidden[-1])
        return out
```

```python
model = LSTMModel(1, 50, 1)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
epochs = 200
```

```python
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    predictions = model(X_tensor)
    print(predictions.shape)
    print(y_tensor.shape)
    break
    loss = criterion(predictions, y_tensor)
    loss.backward()
    optimizer.step()
    if (epoch+1) % 10 == 0:
        print(f"Epochs: {epoch+1} Loss: {loss.item()}")
```

```python
predictions
```

```python
model.eval()
with torch.no_grad():
    scaled_predictions = model(X_tensor).numpy()
```

```python
predictions_original = scaler.inverse_transform(scaled_predictions)
true_values_original = scaler.inverse_transform(y_tensor.numpy())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(true_values_original, label="Actual Values")
plt.plot(predictions_original, label="Predicted Values")
plt.legend()
plt.show()
```
