```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
data = np.array([
    [1, 2],
    [1, 4],
    [1, 0],
    [10, 10],
    [12, 10],
    [10, 12],
    [6, 6],
    [8, 8],
    [7, 8],
    [9, 7]
])
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
def optimize_k_means(data, max_k):
    means = []
    inertias = []
    for i in range(1, max_k):
        kmeans_opt = KMeans(n_clusters=i)
        kmeans_opt.fit(data)
        means.append(i)
        inertias.append(kmeans_opt.inertia_)

    plt.plot(means, inertias, 'o-')
    plt.xlabel("Number of Clusters")
    plt.ylabel("Inertia")
    plt.grid(True)
    plt.show()
```

```python
optimize_k_means(scaled_data, 10)
```

```python
kmeans_optimal = KMeans(n_clusters=2)
kmeans_optimal.fit(scaled_data)
```

```python
labels_optimal = kmeans_optimal.labels_
```

```python
plt.scatter(x=data[:, 0], y=data[:, 1], c=labels_optimal)
```

```python
labels_multi = []
for k in range(1, 6):
    kmeans_multi = KMeans(n_clusters=k)
    kmeans_multi.fit(scaled_data)
    labels_multi.append(kmeans_multi.labels_)
```

```python
fig, axis = plt.subplots(nrows=1, ncols=5, figsize=(20, 5))

for i, ax in enumerate(axis):  # Iterate over `axis`, not `fig.axes`
    ax.scatter(x=data[:, 0], y=data[:, 1], c=labels_multi[i])
    ax.set_title(f"N Clusters: {i + 1}")  # Add 1 to match the cluster count

plt.tight_layout()
plt.show()
```
