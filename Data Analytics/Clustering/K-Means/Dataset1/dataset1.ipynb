```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("force2020_data_unsupervised_learning.csv", index_col="DEPTH_MD")
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.describe()
```

```python
scaler = StandardScaler()
```

```python
df[["RHOB_T", "GR_T", "NPHI_T", "PEF_T", "DTC_T"]] = scaler.fit_transform(df[["RHOB", "GR", "NPHI", "PEF", "DTC"]])
```

```python
df
```

```python
def optimiza_k_means(data, max_k):
    means = []
    inertias = []

    for k in range(1, max_k):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)
        means.append(k)
        inertias.append(kmeans.inertia_)
    
    fig = plt.subplots(figsize=(10, 5))
    plt.plot(means, inertias, 'o-')
    plt.xlabel("Number of Clusters")
    plt.ylabel("Intertia")
    plt.grid(True)
    plt.show()
```

```python
optimiza_k_means(df[["RHOB_T", "NPHI_T"]], 10)
```

```python
kmeans_org = KMeans(n_clusters=3)
kmeans_org.fit(df[["RHOB_T", "NPHI_T"]])
```

```python
df["kmeans_3"] = kmeans_org.labels_
```

```python
df
```

```python
plt.scatter(x=df["RHOB"], y=df["NPHI"], c=df["kmeans_3"])
plt.show()
```

```python
for k in range(1, 6):
    kmeans_diff = KMeans(n_clusters=k)
    kmeans_diff.fit(df[["RHOB_T", "NPHI_T"]])
    df[f"Kmeans-Diff k={k}"] = kmeans_diff.labels_
```

```python
df
```

```python
fig, axis = plt.subplots(nrows=1, ncols=5, figsize=(20,5))

for i, ax in enumerate(fig.axes, start=1):
    ax.scatter(x=df["RHOB_T"], y=df["NPHI_T"], c=df[f"Kmeans-Diff k={i}"])
    ax.set_title(f"N Clusters: {i}")
```
