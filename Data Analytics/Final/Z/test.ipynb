```python
import numpy as np
import pandas as pd
from scipy.stats import multivariate_normal

# Sample DataFrame with delivery locations
data = {
    'Delivery_location_latitude': [40.7128, 40.7130, 40.7125, 40.7132],
    'Delivery_location_longitude': [-74.0060, -74.0058, -74.0062, -74.0059]
}

# Create DataFrame
vrp_df = pd.DataFrame(data)

# Step 1: Calculate the cluster centroid (mean latitude and longitude)
cluster_centroid = vrp_df[["Delivery_location_latitude", "Delivery_location_longitude"]].mean()
#print("Cluster Centroid:")
#print(type(cluster_centroid))

# Step 2: Generate synthetic points around the centroid
def generate_synthetic_points(centroid, spread, num_points):
    print(centroid)
    cov_matrix = np.array([[spread, 0], [0, spread]])  # Covariance matrix (spread in lat/long)
    synthetic_points = multivariate_normal.rvs(mean=centroid, cov=cov_matrix, size=num_points)
    return synthetic_points

# Step 3: Generate synthetic locations around the centroid
spread = 0.001  # Small spread to create nearby points
num_synthetic_points = 5
synthetic_points = generate_synthetic_points(
    centroid=[cluster_centroid["Delivery_location_latitude"], cluster_centroid["Delivery_location_longitude"]],
    spread=spread,
    num_points=num_synthetic_points
)

# Convert the synthetic points into a DataFrame
synthetic_df = pd.DataFrame(synthetic_points, columns=["Delivery_location_latitude", "Delivery_location_longitude"])

#print("\nSynthetic Points Generated Around Centroid:")
#print(synthetic_df)
```
