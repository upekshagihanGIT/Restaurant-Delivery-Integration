```python
# import binomial distribution from scipy
from scipy.stats import binom
```

```python
# number of trials
n = 10
# success probability per trial
p = 0.9
```

```python
# running Binomial Distribution and output
for k in range(n+1):
    probability = binom.pmf(k, n, p)
    print("{0} - {1}".format(k, probability))
```

```python
# function to calculate factorial
def factorial(n: int) -> int:
    f = 1
    for i in range(n , 1, -1):
        f = f * i
    return f
```

```python
# function to calculate binomial coefficient
def binomial_coefficient(n: int, k: int) -> int:
    return factorial(n) / (factorial(n-k) * factorial(k))
```

```python
# define probability mass function
def pmf(p: float, n: int, k: int) -> float:
    return binomial_coefficient(n, k) * (p**k) * (1-p)**(n-k)
```

```python
n = 10
p = 0.9
# since we are successfully identifying 3 bulbs as defective
for k in range(n+1):
    print("Probability of identifying {0} bulbs as defective : ".format(k), pmf(p, n, k))
```
