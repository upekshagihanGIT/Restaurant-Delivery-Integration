```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("final_analysis.csv")
df.head()
```

```python
for col in df.columns:
    if col == "Clustering Method":
        continue
    plt.figure(figsize=(10, 6))
    plt.plot(df["Clustering Method"], df[col], label=col, marker='x', markersize=8)
    plt.xlabel("Clustering Algorithm")
    plt.ylabel("Score")
    plt.legend(loc="best")
    plt.grid(True)
    plt.savefig(f"{col}.png", dpi=300, bbox_inches="tight")
    plt.close()
```

```python
plt.figure(figsize=(10, 6))
for col in df.columns:
    if col == "Clustering Method" or col == "Calinski-Harabasz Index":
        continue
    plt.plot(df["Clustering Method"], df[col], label=col, marker='x', markersize=8)
plt.xlabel("Clustering Algorithm")
plt.ylabel("Score")    
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.savefig(f"aio.png", dpi=300, bbox_inches="tight")
plt.close()
```
