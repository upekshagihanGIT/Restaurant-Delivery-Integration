```python
import random
import csv

import matplotlib.pyplot as plt
```

```python
zones = [
    {"name": "Financial District", "lat_min": 40.700, "lat_max": 40.715, "lon_min": -74.015, "lon_max": -74.000},
    {"name": "SoHo", "lat_min": 40.715, "lat_max": 40.730, "lon_min": -74.010, "lon_max": -73.990},
    {"name": "Chelsea", "lat_min": 40.730, "lat_max": 40.750, "lon_min": -74.005, "lon_max": -73.985},
    {"name": "Midtown", "lat_min": 40.750, "lat_max": 40.770, "lon_min": -73.995, "lon_max": -73.965},
    {"name": "Upper East Side", "lat_min": 40.760, "lat_max": 40.785, "lon_min": -73.975, "lon_max": -73.945},
    {"name": "Greenwich Village", "lat_min": 40.725, "lat_max": 40.740, "lon_min": -74.015, "lon_max": -73.995},
    {"name": "Gramercy", "lat_min": 40.735, "lat_max": 40.750, "lon_min": -73.990, "lon_max": -73.970},
    {"name": "Tribeca", "lat_min": 40.710, "lat_max": 40.725, "lon_min": -74.015, "lon_max": -74.000},
    {"name": "Hell's Kitchen", "lat_min": 40.750, "lat_max": 40.770, "lon_min": -74.005, "lon_max": -73.985},
    {"name": "Upper West Side", "lat_min": 40.770, "lat_max": 40.795, "lon_min": -73.985, "lon_max": -73.955}
]
```

```python
records = []
for _ in range(400):
    # Randomly select a zone
    zone = random.choice(zones)
    # Generate coordinates within the selected zone's boundaries
    lat = random.uniform(zone["lat_min"], zone["lat_max"])
    lon = random.uniform(zone["lon_min"], zone["lon_max"])
    records.append({
        "delivery_location_latitude": round(lat, 4),
        "delivery_location_longitude": round(lon, 4),
        "delivery_zone": zone["name"]
    })
```

```python
with open("delivery_locations.csv", "w", newline="") as csvfile:
    fieldnames = ["delivery_location_latitude", "delivery_location_longitude", "delivery_zone"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for record in records:
        writer.writerow(record)
```
